```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Assuming 'data' has already been loaded
data = pd.read_csv('cleaned_file.csv')

# Filter data for the state "MADHYA PRADESH" and the years 2001 to 2012
df_mp = data[(data['STATE/UT'] == 'MADHYA PRADESH') & (data['YEAR'] >= 2001) & (data['YEAR'] <= 2012)]

# Get the list of crime columns (excluding non-crime columns)
crime_columns = data.columns.difference(['STATE/UT', 'DISTRICT', 'YEAR', 'TOTAL IPC CRIMES'])

# Prepare the figure
plt.figure(figsize=(14, 8))

# Create a bright and visually distinctive color palette
colors = sns.color_palette('tab10', n_colors=len(crime_columns))

# Plot each crime separately using a stacked bar chart
bottom = np.zeros(len(df_mp['YEAR'].unique()))
years = df_mp['YEAR'].unique()
for i, crime_column in enumerate(crime_columns):
    yearly_data = [df_mp[df_mp['YEAR'] == year][crime_column].sum() for year in years]
    plt.bar(years, yearly_data, bottom=bottom, label=crime_column, color=colors[i], edgecolor='white')
    bottom += np.array(yearly_data)

plt.title('Crime Trends in MADHYA PRADESH (2001-2012)')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Type', loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, ncol=4)
plt.xticks(range(2001, 2013), rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

# Load the dataset
data = pd.read_csv('cleaned_file.csv')  # Ensure you load your actual data path

# Sum total IPC crimes per state
state_total_crimes = data.groupby('STATE/UT')['TOTAL IPC CRIMES'].sum().reset_index()

# Define crime rate categories based on quantiles
quantiles = state_total_crimes['TOTAL IPC CRIMES'].quantile([0.33, 0.66])
low_threshold, high_threshold = quantiles.values

# Categorize the total crime rates
state_total_crimes['Crime Category'] = pd.cut(state_total_crimes['TOTAL IPC CRIMES'],
                                              bins=[0, low_threshold, high_threshold, float('inf')],
                                              labels=['Low', 'Medium', 'High'])

# Prepare data for the mosaic plot
crime_category_counts = state_total_crimes.groupby(['STATE/UT', 'Crime Category']).size().unstack(fill_value=0)
crime_category_props = crime_category_counts.div(crime_category_counts.sum(axis=1), axis=0)

# Create the mosaic plot
plt.figure(figsize=(20, 12))  # Increased figure size for better clarity
mosaic(crime_category_props.stack(), gap=0.05, title='State/UT vs. Crime Category Proportions')  # Increased gap size
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
#to perform bootstrap
def bootstrap(data, num_iterations=2000):
    bootstrap_means = []
    for _ in range(num_iterations):
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        bootstrap_mean = np.mean(bootstrap_sample)
        bootstrap_means.append(bootstrap_mean)
    return bootstrap_means

bootstrap_means = bootstrap(total_ipc_crimes)

# Plot
sns.histplot(bootstrap_means, kde=True)
plt.title('Bootstrap Distribution of Sample Means')
plt.xlabel('Bootstrap Sample Mean Total IPC Crimes')
plt.ylabel('Frequency')
plt.show()
```

```python
# the 95% confidence interval
confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])
print("95% Confidence Interval :", confidence_interval)

# Plot
sns.histplot(bootstrap_means, kde=True)
plt.axvline(confidence_interval[0], color='red', linestyle='--', label='95% CI Lower Bound')
plt.axvline(confidence_interval[1], color='red', linestyle='--', label='95% CI Upper Bound')

# Adding x and y labels
plt.xlabel('Sample Mean of Total IPC Crimes')  # X label
plt.ylabel('Density')  # Y label

plt.legend()
plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load the data
data = pd.read_csv('/Users/yashsingh/Desktop/DAV_Project_Crime/cleaned_file.csv')

# Filter data for the years 2005-2007 and 2008-2010
early_years = data[data['YEAR'].isin([2005, 2006, 2007])]['THEFT']
later_years = data[data['YEAR'].isin([2008, 2009, 2010])]['THEFT']

# Perform the t-test
t_stat, p_value = ttest_ind(early_years, later_years, equal_var=False)

# Print the results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpretation of the t-test results
if p_value < 0.05:
    print("There is sufficient statistical evidence to reject the null hypothesis. This suggests that there has been a significant change in the mean number of theft cases from the early years to the later years.")
else:
    print("There is not enough statistical evidence to reject the null hypothesis. This suggests that the mean number of theft cases has not significantly changed from the early years to the later years.")
```

```python
import pandas as pd
from scipy.stats import f_oneway

# Load the data
data = pd.read_csv('/Users/yashsingh/Desktop/DAV_Project_Crime/cleaned_file.csv')

# Collect theft data by year
theft_data = {year: data[data['YEAR'] == year]['THEFT'] for year in data['YEAR'].unique()}

# Perform ANOVA
f_stat, p_value = f_oneway(*theft_data.values())

# Print the results
print("F-statistic:", f_stat)
print("P-value:", p_value)

# Interpretation of the ANOVA results
if p_value < 0.05:
    print("There is sufficient statistical evidence to reject the null hypothesis. This suggests that there are significant differences in the average number of theft cases across different years.")
else:
    print("There is not enough statistical evidence to reject the null hypothesis. This suggests that there are no significant differences in the average number of theft cases across different years.")
```
